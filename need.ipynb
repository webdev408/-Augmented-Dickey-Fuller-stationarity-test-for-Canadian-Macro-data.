```python
diff_df = numeric_cols.copy()
diff_df['d_gdp_billions'] = diff_df['gdp_billions'].diff()
diff_df['d_pop_millions'] = diff_df['pop_millions'].diff()
diff_df['d_longevity'] = diff_df['longevity'].diff()
diff_df['d_mean_taxRate'] = diff_df['mean_taxRate'].diff()
clean_df = diff_df.dropna()
# run the difference regression
diff_formula = 'd_gdp_billions~d_pop_millions+longevity+mean_taxRate'
diff_reg = smf.ols(diff_formula, data=clean_df).fit()
display(diff_reg.summary())
```

```python
from statsmodels.tsa.stattools import adfuller

def adf_test(series, name, maxlag=None):
  result = adfuller(series.dropna(), maxlag=2)
  return{
	'Variable': name,
	'ADF Statistic': result[0],
	'P-Value': result[1],
	'Lags Used': result[2], # this shows manual lag chosen not AIC which choses automatically
	'Observations': result[3],
# Critical value reference inside result[4]
	'Critical 1%': result[4]['1%'],
	'Critical 5%': result[4]['5%'],
	'Critical 10%': result[4]['10%'],
	'Stationary': 'Yes' if result[1] < 0.05 else 'No'
	}
variables_to_test = ['d_pop_millions', 'd_longevity', 'd_mean_taxRate']
# Build the list of result first then create dataframe
adf_result = [adf_test(clean_df[var], var) for var in variables_to_test]
adf_df = pd.DataFrame(adf_result)
# Print results
print("Augmented Dickey Fuller Test for Stationarity")
print("\nHo:Series has a unit root(non-stationary)")
print("H1:Series is stationary")
print("\nNote: 'Lags Used' are manually selected not AIC selected")
# Display results
print(adf_df[['Variable','ADF Statistic','P-Value','Lags Used','Stationary']])
print('--'*60)
print("Critical Value Reference: ")
print(adf_df[['Variable', 'Critical 1%','Critical 5%','Critical 10%']].to_string(index=False))
print("\nInterpretation: if ADF Statistic < Critical Value -> reject Ho -> Stationary")
```

## ARCH/GARCH Model Analysis

Your observations are correct - autocorrelation and heteroscedasticity violations don't necessarily hurt prediction quality, but they do violate BLUE (Best Linear Unbiased Estimator) properties, specifically **efficiency**. The predictions remain unbiased but standard errors are incorrect, affecting inference.

Let's test for ARCH effects (conditional heteroscedasticity) in the residuals:

```python
# First, let's examine the residuals from our differenced model
residuals = y - y_pred

# Plot residuals and squared residuals over time
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

# Residuals plot
ax1.plot(residuals.index, residuals, 'b-', alpha=0.7)
ax1.axhline(y=0, color='r', linestyle='--')
ax1.set_title('Residuals from Differenced Model')
ax1.set_ylabel('Residuals')
ax1.grid(True)

# Squared residuals (proxy for volatility)
ax2.plot(residuals.index, residuals**2, 'g-', alpha=0.7)
ax2.set_title('Squared Residuals (Volatility Proxy)')
ax2.set_xlabel('Time Period')
ax2.set_ylabel('Squared Residuals')
ax2.grid(True)

plt.tight_layout()
plt.show()

print("Visual inspection: clusters of high/low squared residuals suggest ARCH effects")
```

```python
# Test for ARCH effects using Engle's ARCH test (Lagrange Multiplier test)
from statsmodels.stats.diagnostic import het_arch

# Test for ARCH effects with different lags
print("=" * 70)
print("ENGLE'S ARCH TEST (Lagrange Multiplier Test)")
print("=" * 70)
print("H0: No ARCH effects (homoscedastic residuals)")
print("H1: ARCH effects present (conditional heteroscedasticity)\n")

for lag in [1, 2, 3, 4]:
    lm_stat, lm_pvalue, f_stat, f_pvalue = het_arch(residuals, nlags=lag)
    print(f"ARCH({lag}) Test:")
    print(f"  LM Statistic: {lm_stat:.4f}, P-value: {lm_pvalue:.4f}")
    print(f"  F-Statistic: {f_stat:.4f}, P-value: {f_pvalue:.4f}")
    print(f"  Result: {'ARCH effects detected' if lm_pvalue < 0.05 else 'No significant ARCH effects'} (=0.05)\n")
```

```python
# Install arch package if needed (uncomment if not installed)
# !pip install arch

from arch import arch_model

# Prepare data: ARCH models require mean-zero returns
# We'll use the residuals from our linear model
residuals_series = pd.Series(residuals.values, index=residuals.index)

# Specify and fit ARCH(1) model
print("=" * 70)
print("FITTING ARCH(1) MODEL")
print("=" * 70)
arch1_model = arch_model(residuals_series, vol='ARCH', p=1, mean='Zero')
arch1_fit = arch1_model.fit(disp='off')
print(arch1_fit.summary())
print("\n")
```

```python
# Fit GARCH(1,1) model - most commonly used specification
print("=" * 70)
print("FITTING GARCH(1,1) MODEL")
print("=" * 70)
garch11_model = arch_model(residuals_series, vol='GARCH', p=1, q=1, mean='Zero')
garch11_fit = garch11_model.fit(disp='off')
print(garch11_fit.summary())
print("\n")

# Compare model fit using information criteria
print("=" * 70)
print("MODEL COMPARISON")
print("=" * 70)
comparison = pd.DataFrame({
    'Model': ['ARCH(1)', 'GARCH(1,1)'],
    'AIC': [arch1_fit.aic, garch11_fit.aic],
    'BIC': [arch1_fit.bic, garch11_fit.bic],
    'Log-Likelihood': [arch1_fit.loglikelihood, garch11_fit.loglikelihood]
})
print(comparison.to_string(index=False))
print("\nLower AIC/BIC indicates better model fit")
```

```python
# Visualize conditional volatility from GARCH model
fig, axes = plt.subplots(3, 1, figsize=(14, 10))

# Original residuals
axes[0].plot(residuals.index, residuals, 'b-', alpha=0.7, label='Residuals')
axes[0].axhline(y=0, color='r', linestyle='--', alpha=0.5)
axes[0].set_title('Model Residuals', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Residuals')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Conditional volatility (standard deviation)
garch_vol = garch11_fit.conditional_volatility
axes[1].plot(residuals.index, garch_vol, 'g-', linewidth=2, label='Conditional Volatility ()')
axes[1].set_title('GARCH(1,1) Conditional Volatility', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Volatility ()')
axes[1].legend()
axes[1].grid(True, alpha=0.3)
axes[1].fill_between(residuals.index, 0, garch_vol, alpha=0.3, color='green')

# Standardized residuals
std_resid = garch11_fit.std_resid
axes[2].plot(residuals.index, std_resid, 'purple', alpha=0.7, label='Standardized Residuals')
axes[2].axhline(y=0, color='r', linestyle='--', alpha=0.5)
axes[2].axhline(y=2, color='orange', linestyle=':', alpha=0.5, label='2 threshold')
axes[2].axhline(y=-2, color='orange', linestyle=':', alpha=0.5)
axes[2].set_title('Standardized Residuals (should be homoscedastic)', fontsize=12, fontweight='bold')
axes[2].set_xlabel('Time Period')
axes[2].set_ylabel('Std. Residuals')
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

   - Risk management insights

## Trend Stationary (TS) vs Difference Stationary (DS)

### Important Clarification:

**You have it backwards!** Let me correct this:

- **Differencing** corrects **Difference Stationary (DS)** processes (stochastic trend/unit root)
- **Detrending** corrects **Trend Stationary (TS)** processes (deterministic trend)

### The Two Types:

1. **Trend Stationary (TS)**: $y_t = \alpha + \beta t + \varepsilon_t$
   - Deterministic trend (predictable time trend)
   - Shocks have temporary effects
   - **Fix**: Detrend (regress on time, use residuals)

2. **Difference Stationary (DS)**: $y_t = y_{t-1} + \varepsilon_t$ (random walk)
   - Stochastic trend (unit root)
   - Shocks have permanent effects
   - **Fix**: Difference (take $\Delta y_t = y_t - y_{t-1}$)

Let's test which one your data exhibits:

```python
# Test original series for unit roots (DS behavior)
print("=" * 80)
print("UNIT ROOT TESTS ON ORIGINAL (LEVEL) SERIES")
print("=" * 80)
print("H0: Series has a unit root (DS - Difference Stationary)")
print("H1: Series is stationary (or TS - Trend Stationary)\n")

original_vars = ['gdp_billions', 'pop_millions', 'longevity']
level_results = []

for var in original_vars:
    result = adfuller(df[var].dropna(), maxlag=2)
    level_results.append({
        'Variable': var,
        'ADF Stat': result[0],
        'P-Value': result[1],
        'Critical 5%': result[4]['5%'],
        'Conclusion': 'Stationary' if result[1] < 0.05 else 'Unit Root (DS)'
    })

level_df = pd.DataFrame(level_results)
print(level_df.to_string(index=False))
print("\nInterpretation: P-value > 0.05  Fail to reject H0  Unit Root  DS process")
print("                P-value < 0.05  Reject H0  Stationary (possibly TS)")
print("\n" + "=" * 80)
```

```python
# Compare: Detrending (for TS) vs Differencing (for DS)
print("=" * 80)
print("COMPARISON: DETRENDING vs DIFFERENCING")
print("=" * 80)

# Method 1: Detrending (assumes TS)
from scipy import stats
time_index = np.arange(len(df))
gdp_detrended = df['gdp_billions'] - stats.linregress(time_index, df['gdp_billions']).slope * time_index - stats.linregress(time_index, df['gdp_billions']).intercept

# Method 2: Differencing (assumes DS) - already have this
gdp_differenced = df['gdp_billions'].diff().dropna()

# Test both transformed series
detrend_adf = adfuller(gdp_detrended.dropna(), maxlag=2)
diff_adf = adfuller(gdp_differenced.dropna(), maxlag=2)

comparison_results = pd.DataFrame({
    'Method': ['Original GDP', 'Detrended (TS fix)', 'Differenced (DS fix)'],
    'ADF Statistic': [level_results[0]['ADF Stat'], detrend_adf[0], diff_adf[0]],
    'P-Value': [level_results[0]['P-Value'], detrend_adf[1], diff_adf[1]],
    'Stationary?': [
        'No' if level_results[0]['P-Value'] > 0.05 else 'Yes',
        'Yes' if detrend_adf[1] < 0.05 else 'No',
        'Yes' if diff_adf[1] < 0.05 else 'No'
    ]
})

print(comparison_results.to_string(index=False))
print("\n" + "=" * 80)
```

```python
# Visualize the difference between TS and DS
fig, axes = plt.subplots(3, 1, figsize=(14, 10))

# Original series
axes[0].plot(df.index, df['gdp_billions'], 'b-', linewidth=2, label='Original GDP')
axes[0].set_title('Original GDP Series (Non-Stationary)', fontsize=12, fontweight='bold')
axes[0].set_ylabel('GDP (billions)')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Detrended (TS correction)
axes[1].plot(df.index, gdp_detrended, 'g-', alpha=0.7, label='Detrended (TS fix)')
axes[1].axhline(y=0, color='r', linestyle='--', alpha=0.5)
axes[1].set_title(f'Detrended GDP (P-value: {detrend_adf[1]:.4f})', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Detrended GDP')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

# Differenced (DS correction)
axes[2].plot(gdp_differenced.index, gdp_differenced, 'purple', alpha=0.7, label='Differenced (DS fix)')
axes[2].axhline(y=0, color='r', linestyle='--', alpha=0.5)
axes[2].set_title(f'Differenced GDP (P-value: {diff_adf[1]:.4f})', fontsize=12, fontweight='bold')
axes[2].set_xlabel('Year')
axes[2].set_ylabel('GDP')
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nVisual Comparison:")
print("- Detrended: Removes linear trend, but still shows some non-stationarity")
print("- Differenced: Removes the stochastic trend, achieves stationarity")
```
